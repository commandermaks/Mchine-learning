```python
df_temp.set_index('name', inplace=True)
```

```python
X=df_temp.drop(['party','state'], axis = 1)
```

```python
Y=df_temp['party']
```

```python
ward = AgglomerativeClustering(n_clusters=3)
ward_pred = ward.fit_predict(X)
```

```python
from scipy.cluster.hierarchy import linkage

#Write code here
linkage_type = 'complete'

linkage_matrix = linkage(X, linkage_type)
```

```python
import seaborn as sns
lut ={'D':'b', 'R':'g', 'I':'r'}
row_colors = Y.map(lut)
sns.clustermap(X, row_colors=row_colors, figsize=(12,10),method='complete')
plt.show()
```
