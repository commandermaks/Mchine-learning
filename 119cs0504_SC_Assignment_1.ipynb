```python
import numpy as np
import skfuzzy as fuzz
from skfuzzy import control as ctrl
```

```python
marketvalue = ctrl.Antecedent(np.arange(0, 101, 1),'marketvalue')
location = ctrl.Antecedent(np.arange(0, 11, 1), 'location')
house = ctrl.Consequent(np.arange(0, 11, 1), 'house')

marketvalue['low'] = fuzz.trapmf(marketvalue.universe, [0, 0,5,10])
marketvalue['medium'] = fuzz.trapmf(marketvalue.universe, [5,10,20,25])
marketvalue['high'] = fuzz.trapmf(marketvalue.universe, [20,25,65,85])
marketvalue['veryhigh'] = fuzz.trapmf(marketvalue.universe, [65,85,100,100])

```

```python
marketvalue.view()
```

```python
location['bad'] = fuzz.trapmf(location.universe, [0, 0, 2,4])
location['fair'] = fuzz.trapmf(location.universe, [3,5,6,9])
location['excellent'] = fuzz.trapmf(location.universe, [6,9,10,10])

location.view()
```

```python
house['verylow'] = fuzz.trimf(house.universe, [0, 0, 3])
house['low'] = fuzz.trimf(house.universe, [0, 3, 6])
house['medium'] = fuzz.trimf(house.universe, [2, 5, 8])
house['high'] = fuzz.trimf(house.universe, [4, 7, 10])
house['veryhigh'] = fuzz.trimf(house.universe, [7, 10, 10])
house.view()
```

```python
rule1 = ctrl.Rule(marketvalue['low'] , house['low'])
rule2 = ctrl.Rule(location['bad'] ,house['low'])
rule3 = ctrl.Rule(location['bad'] and marketvalue['low'], house['verylow'])
rule4 = ctrl.Rule(location['bad'] and marketvalue['medium'], house['low'])
rule5 = ctrl.Rule(location['bad'] and marketvalue['high'], house['medium'])
rule6 = ctrl.Rule(location['bad'] and marketvalue['veryhigh'], house['high'])
rule7 = ctrl.Rule(location['fair'] and marketvalue['low'], house['low'])
rule8 = ctrl.Rule(location['fair'] and marketvalue['medium'], house['medium'])
rule9 = ctrl.Rule(location['fair'] and marketvalue['high'], house['high'])
rule10 = ctrl.Rule(location['fair'] and marketvalue['veryhigh'], house['veryhigh'])
rule11 = ctrl.Rule(location['excellent'] and marketvalue['low'], house['medium'])
rule12 = ctrl.Rule(location['excellent'] and marketvalue['medium'], house['high'])
rule13 = ctrl.Rule(location['excellent'] and marketvalue['high'], house['veryhigh'])
rule14 = ctrl.Rule(location['excellent'] and marketvalue['veryhigh'], house['veryhigh'])

rule1.view()
rule2.view()
rule3.view()
rule4.view()
rule5.view()
rule6.view()
rule7.view()
rule8.view()
rule9.view()
rule10.view()
rule11.view()
rule12.view()
rule13.view()
rule14.view()



```
